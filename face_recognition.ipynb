```python
for dir_path,dir_name,filename in walk(r"C:\Users\archa\OneDrive\Desktop\Downloads\att_faces-20240430T052931Z-001\att_faces"):
    for fn in filename:
        if fn[-3:]=='pgm':
            image_filename=path.join(dir_path,fn)
            print(image_filename)
        image=cv2.imread(image_filename,cv2.IMREAD_GRAYSCALE)
        scaled_img=scale(image.reshape(10304).astype('float32')) 
        x.append(scaled_img)
        y.append(dir_path)
```

```python
x
```

```python
x=np.array(x)
x.shape
```

```python
import matplotlib.pyplot as plt
xtrain,xtest,ytrain,ytest=train_test_split(x,y)
print(xtrain.shape)
pca=PCA()
reduced_x=pca.fit_transform(xtrain)
print(reduced_x.shape)
```

```python
pca.n_components_
```

```python
pca_values=np.arange(pca.n_components_)
pca_values
```

```python
pca.explained_variance_ratio_
```

```python
plt.plot(pca_values,pca.explained_variance_ratio_,'o-',linewidth=2,color='blue')
plt.title('screeplot')
plt.xlabel('principal component')
plt.ylabel('eigen value')
plt.show()
```

```python
pca=PCA(n_components=200)
reduced_x=pca.fit_transform(xtrain)
x_test_red=pca.transform(xtest)
```

```python
classifier=LogisticRegression()
classifier.fit(reduced_x,ytrain)
```

```python
prediction=classifier.predict(x_test_red)
prediction[2]
```

```python
print(classification_report(ytest,prediction))
```
